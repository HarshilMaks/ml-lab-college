```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 1. Data Generate kiya (100 points)
data = np.random.rand(100)

# 2. Pehle 50 points ko label kiya (Training set)
train_data = data[:50]
train_labels = ["Class1" if x <= 0.5 else "Class2" for x in train_data]

# Baki 50 testing ke liye
test_data = data[50:]

# 3. Distance function
def euclidean_distance(x1, x2):
    return abs(x1 - x2)

# 4. k-NN Classifier
def knn_classifier(train_data, train_labels, test_point, k):
    distances = []
    for i in range(len(train_data)):
        dist = euclidean_distance(test_point, train_data[i])
        distances.append((dist, train_labels[i]))
    
    # Distance ke basis pe sort kiya (x[0] is distance)
    distances.sort(key=lambda x: x[0])
    
    # Top k neighbors uthaye
    k_nearest_labels = [label for _, label in distances[:k]]
    
    # Majority vote (Fix: [0][0] lagaya taaki string mile)
    return Counter(k_nearest_labels).most_common(1)[0][0]

# 5. Testing, Calculations Print aur Visualization
k_values = [1, 3, 5, 20]
print("----- k-NN Classification Results -----\n")

for k in k_values:
    print(f"\nResults for k = {k}:")
    classified_labels = []
    
    for i, tp in enumerate(test_data):
        label = knn_classifier(train_data, train_labels, tp, k)
        classified_labels.append(label)
        # Ye rahi teri calculations print wali line
        print(f"Point x{i+51} (value: {tp:.4f}) -> {label}")
    
    # Plotting
    plt.figure(figsize=(10, 4))
    
    # Training data points (Bottom row)
    train_colors = ["blue" if l == "Class1" else "red" for l in train_labels]
    plt.scatter(train_data, [0] * len(train_data), c=train_colors, label="Train Data", alpha=0.5)
    
    # Predicted test points (Top row)
    test_colors = ["blue" if l == "Class1" else "red" for l in classified_labels]
    plt.scatter(test_data, [1] * len(test_data), c=test_colors, marker="x", label=f"Test (k={k})")
    
    plt.title(f"k-NN Classification for k = {k}")
    plt.yticks([0, 1], ["Train", "Test"])
    plt.legend()
    plt.show()

print("\nClassification Complete.")
```
