```python
import pandas as pd

def find_s_algorithm(file_path):
    # Read the dataset
    data = pd.read_csv(file_path)
    print("Training data:")
    print(data)

    # Separate attributes and target
    attributes = data.iloc[:, :-1]
    target = data.iloc[:, -1]

    # Initialize hypothesis with the first positive example
    # Find the index of the first 'Yes'
    first_pos_idx = target[target == 'Yes'].index[0]
    hypothesis = list(attributes.iloc[first_pos_idx])

    print(f"\nInitial specific hypothesis: {hypothesis}")

    # Iterate through the dataset
    for i, row in data.iterrows():
        if target[i] == 'Yes':
            for j in range(len(hypothesis)):
                if row.iloc[j] != hypothesis[j]:
                    # Generalize if values don't match
                    hypothesis[j] = '?'
                else:
                    pass
        
    return hypothesis

# Path to your file
file_path = r"C:\Users\AMC College\Downloads\training_data.csv"
# Run the algorithm
try:
    final_h = find_s_algorithm(file_path)
    print("\nThe final specific hypothesis is:", final_h)
except Exception as e:
    print(f"Error: {e}. Check if your CSV path is correct and has 'Yes' labels.")
```
