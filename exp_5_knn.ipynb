# k-Nearest Neighbors (k-NN) Classification Example

This notebook demonstrates a simple k-NN classifier on randomly generated data, with step-by-step explanations and visualizations.

```python
# Import required libraries for k-NN classification and visualization
import numpy as np  # For numerical operations and random data generation
import matplotlib.pyplot as plt  # For plotting graphs
from collections import Counter  # For majority voting in k-NN
```

## Generate Data

Create 100 random data points and split them into training and testing sets.

```python
# Generate 100 random data points between 0 and 1
# These points will be used for both training and testing
# The first 50 points are for training, the rest for testing

data = np.random.rand(100)  # Generate 100 random floats in [0, 1)

# First 50 points for training
train_data = data[:50]  # Training data
# Assign class labels: <=0.5 is 'Class1', >0.5 is 'Class2'
train_labels = ["Class1" if x <= 0.5 else "Class2" for x in train_data]

# Remaining 50 for testing
test_data = data[50:]  # Testing data
```

## Define Distance Function

Define a function to compute the Euclidean distance between two points.

```python
# Euclidean distance function for 1D points
# Returns the absolute difference between two points
# Used to measure similarity in k-NN

def euclidean_distance(x1, x2):
    return abs(x1 - x2)
```

## k-NN Classifier Function

Define the k-NN classifier that predicts the class of a test point.

```python
# k-NN classifier implementation for 1D data
# Given a test point, finds the k nearest neighbors in the training set
# and returns the most common class label among them

def knn_classifier(train_data, train_labels, test_point, k):
    distances = []  # List to store (distance, label) tuples
    for i in range(len(train_data)):
        dist = euclidean_distance(test_point, train_data[i])  # Compute distance
        distances.append((dist, train_labels[i]))  # Store distance and label
    # Sort by distance (ascending)
    distances.sort(key=lambda x: x[0])
    # Get top k neighbors' labels
    k_nearest_labels = [label for _, label in distances[:k]]
    # Majority vote: return the most common label among neighbors
    return Counter(k_nearest_labels).most_common(1)[0][0]
```

## Test and Visualize k-NN Results

Classify test points for different values of k and visualize the results.

```python
# Test the classifier for different k values and visualize the results
k_values = [1, 3, 5, 20]  # Different values of k to test
print("----- k-NN Classification Results -----\n")
for k in k_values:
    print(f"\nResults for k = {k}:")
    classified_labels = []  # Store predicted labels for test data
    for i, tp in enumerate(test_data):
        label = knn_classifier(train_data, train_labels, tp, k)  # Predict label
        classified_labels.append(label)
        # Print the classification result for each test point
        print(f"Point x{i+51} (value: {tp:.4f}) -> {label}")
    # Plotting results
    plt.figure(figsize=(10, 4))
    # Training data points (bottom row, colored by class)
    train_colors = ["blue" if l == "Class1" else "red" for l in train_labels]
    plt.scatter(train_data, [0] * len(train_data), c=train_colors, label="Train Data", alpha=0.5)
    # Predicted test points (top row, colored by predicted class)
    test_colors = ["blue" if l == "Class1" else "red" for l in classified_labels]
    plt.scatter(test_data, [1] * len(test_data), c=test_colors, marker="x", label=f"Test (k={k})")
    plt.title(f"k-NN Classification for k = {k}")
    plt.yticks([0, 1], ["Train", "Test"])
    plt.legend()
    plt.show()
print("\nClassification Complete.")
```
