# Find-S Algorithm Example

This notebook demonstrates the Find-S algorithm for learning the most specific hypothesis from positive examples in a dataset. Each step is explained for clarity.

```python
# Import required library for data handling
import pandas as pd  # For reading CSV and manipulating data
```

## Define the Find-S Algorithm

This function implements the Find-S algorithm to find the most specific hypothesis from positive examples.

```python
def find_s_algorithm(file_path):
    # Read the dataset from the given CSV file path
    data = pd.read_csv(file_path)
    print("Training data:")
    print(data)

    # Separate attributes (features) and target (class label)
    attributes = data.iloc[:, :-1]  # All columns except last
    target = data.iloc[:, -1]       # Last column (target)

    # Initialize hypothesis with the first positive example (target == 'Yes')
    # Find the index of the first positive example
    first_pos_idx = target[target == 'Yes'].index[0]
    hypothesis = list(attributes.iloc[first_pos_idx])  # Start with this example

    print(f"\nInitial specific hypothesis: {hypothesis}")

    # Iterate through all examples in the dataset
    for i, row in data.iterrows():
        if target[i] == 'Yes':  # Only consider positive examples
            for j in range(len(hypothesis)):
                if row.iloc[j] != hypothesis[j]:
                    # If attribute value differs, generalize with '?'
                    hypothesis[j] = '?'
    return hypothesis  # Return the most specific hypothesis found
```

## Run the Algorithm

Specify the path to your CSV file and run the Find-S algorithm. The final hypothesis will be printed.

```python
# Path to your file (update this to your actual CSV location)
file_path = r"training_data.csv"
# Run the Find-S algorithm and print the result
try:
    final_h = find_s_algorithm(file_path)  # Run the algorithm
    print("\nThe final specific hypothesis is:", final_h)  # Output the result
except Exception as e:
    # Print error if file not found or data is not as expected
    print(f"Error: {e}. Check if your CSV path is correct and has 'Yes' labels.")
```
